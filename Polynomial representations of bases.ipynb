```python
# Define function to generate the coefficients of a polynomial for any number in any base
def getCoeffs(base, number):
    coeffs = {}
    i = 0
    c = 0
    x = number
    exp = 0
    
    # Calculate the power
    while(True):
        i+=1
        next_exp = base**i
        if(next_exp > number):
            i = i-1
            break
        exp = next_exp
        
    # Calculate the coefficient
    while(True):
        c+=1
        next_coeff = exp*c
        if(next_coeff > number):
            c = c-1
            break
        coeff = next_coeff
        
        
    coeffs[i] = c
    r = number - coeff
    
    if r >= base: 
        coeffs.update(getCoeffs(base, r))
    else:
        if r != 0: coeffs[0] = r
    return coeffs
    
    
```
